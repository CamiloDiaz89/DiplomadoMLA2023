### Ejercicio 5 - Aglomerativo: Se comparan los linkage (complete, single, average, ward)

#### Importar las librerias

```python
# Libreria para las matematicas
import numpy as np
# Libreria para el el manejo de los datos
import pandas as pd
# Libreria para gráficas
import matplotlib.pyplot as plt
# Libreria para el conjunto de datos
from sklearn import datasets
# Libreria para el dendograma
import scipy.cluster.hierarchy as sch
# Libreria para el modelo Aglomerativo
from sklearn.cluster import AgglomerativeClustering
# Libreria para la métrica
from sklearn.metrics import adjusted_rand_score
```

#### Cargue de datos desde el paquete dataset

```python
iris = datasets.load_iris()
X, y = iris.data[:, [2,3]], iris.target

print("Caracteristicas: ", iris.feature_names)
print("Objetivo: ", iris.target_names)
```

#### Visualización de los datos

```python
plt.figure(figsize=(10,7))
plt.scatter(X[:,0], X[:,1], c = y)
plt.title('DataSet Iris')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```

#### Complete

```python
### 1- Dendrograma
plt.figure(figsize=(10,7))
cluster = sch.dendrogram(sch.linkage(X, method="complete"))
plt.axhline(y=5, c="r", linestyle="--")
```

```python
### 2- Predicción
ac = AgglomerativeClustering(n_clusters=3, linkage='complete')
prediccion_complete = ac.fit_predict(X)
prediccion_complete
```

```python
#### 3- Métrica
print("Rand Score Complete:", round(adjusted_rand_score(y, prediccion_complete),2))
```

```python
### 4- Gráfica comparativa de valores reales VS predicción
plt.figure(figsize=(15, 8))

# Datos originales
plt.subplot(1,2,1)
plt.title('Datos originales')
plt.scatter(X[y==0,0], X[y==0,1], c='red', marker='o')
plt.scatter(X[y==1,0], X[y==1,1], c='blue', marker='o')
plt.scatter(X[y==2,0], X[y==2,1], c='green', marker='o')

# Datos de la predicción
plt.subplot(1,2,2)
plt.title('Datos de la predicción')
plt.scatter(X[prediccion_complete==0,0], X[prediccion_complete==0,1], c='red', marker='o')
plt.scatter(X[prediccion_complete==1,0], X[prediccion_complete==1,1], c='blue', marker='o')
plt.scatter(X[prediccion_complete==2,0], X[prediccion_complete==2,1], c='green', marker='o')

plt.show()
```

#### Single

```python
### 1- Dendrograma
plt.figure(figsize=(10,7))
cluster = sch.dendrogram(sch.linkage(X, method="single"))
plt.axhline(y=0.4, c="r", linestyle="--")
```

```python
### 2- Predicción
ac = AgglomerativeClustering(n_clusters=3, linkage='single')
prediccion_single = ac.fit_predict(X)
prediccion_single
```

```python
#### 3- Métrica
print("Rand Score Single:", round(adjusted_rand_score(y, prediccion_single),2))
```
